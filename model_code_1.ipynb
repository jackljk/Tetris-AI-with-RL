```python
from Custom_Tetris_Env.tetris import TetrisEnv
from Custom_Tetris_Env.jp import JoypadSpace
from Custom_Tetris_Env.actions import MOVEMENT
from stable_baselines3 import DQN
from stable_baselines3.common.callbacks import CheckpointCallback



checkpoint_callback = CheckpointCallback(save_freq=100000, save_path='./logs/',
                                         name_prefix='dqn_model')

env = TetrisEnv(reward_lines=True, reward_score=True, penalize_height=True)
env = JoypadSpace(env, MOVEMENT)

model = DQN("MlpPolicy", env, verbose=1)
```

```python
# Train the agent
model.learn(total_timesteps=int(1e5), callback=checkpoint_callback)
```

```python
env = TetrisEnv()
env = JoypadSpace(env, MOVEMENT)

# load checkpoint
model = DQN.load("logs/dqn_model_1000000_steps.zip", env=env)
```

```python
# continue training
model.learn(total_timesteps=int(1e6), callback=checkpoint_callback)
```

```python
done = False
obs = env.reset()


while not done:
    action, _states = model.predict(obs)
    obs, reward, done, info = env.step(action)
    env.render()
```
